```python
from matplotlib import pyplot as plt
import seaborn as sns
```

# comparision when step = 200, training data = 400
training size = 400, 
step size = 200, 
batch size = 200, 
test size = 100
total data = 1600

```python
AL_list= [0.44, 0.62, 0.79, 0.82, 0.8, 0.8, 0.89]
```

```python
random_list = [0.44, 0.52, 0.69, 0.49, 0.48, 0.53, 0.66]
```

```python
# Comparision when train size = 1600, step size=500, batch size = 200, test size =100
sns.set_style("whitegrid")
plt.figure(figsize=(15,8))
sns.pointplot(x=range(400,1800,200), y=AL_list,linestyles="-",color="green")
plt.legend("A", ncol=2, loc='upper left');
sns.pointplot(x=range(400,1800,200), y=random_list,linestyles="-")

plt.xlabel("samples")
plt.ylabel("Accuracy")
plt.grid(True)

```
